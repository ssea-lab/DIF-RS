```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import LogNorm

%matplotlib inline
sns.set(font_scale=1.5)
```

```python
#
gapdata = np.loadtxt('/ssd_data/lzw/DIF-RS/models/heatmap/gap.txt')
row_labels = ['1', '2', '4', '8', '16','32']
column_labels = ['1', '2', '5', '10', '20','50']
df = pd.DataFrame(gapdata, columns=column_labels, index=row_labels)
sns.set_context({"figure.figsize":(10,10)})

sns.heatmap(data=df,square=True,cmap="YlGnBu",annot=True,fmt=".3f") 
#
#
#  x  y 
plt.xlabel('Inference diffusion steps')
plt.ylabel('Samples')


plt.title('Gap(%)')
plt.show()
```

# efficiency

```python
#
gapdata = np.loadtxt('/ssd_data/lzw/DIF-RS/models/heatmap/efficiency.txt')
row_labels = ['1', '2', '4', '8', '16','32']
column_labels = ['1', '2', '5', '10', '20','50']
df = pd.DataFrame(gapdata, columns=column_labels, index=row_labels)
sns.set_context({"figure.figsize":(10,10)})

sns.heatmap(data=df,square=True,cmap="YlGnBu",annot=True,fmt=".3f",norm=LogNorm(vmin=0.01, vmax=40)) 
#
#
#  x  y 
plt.xlabel('Inference diffusion steps')
plt.ylabel('Samples')


plt.title('Efficiency(seconds)')
plt.show()
```
