```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# 下面是resource_utilization折线对比图

```python
# 定义列名
column_names = ['1', '2', '4', '8', '16', 'Algorithm']

# 读取数据文件
df = pd.read_csv('/ssd_data/lzw/DIF-RS/models/robust/resource_utilization.txt', sep=' ', names=column_names)

# 打印读取的数据以验证
print(df)

# 将数据转换为适合绘图的格式
df_melted = df.melt(id_vars='Algorithm', var_name='Metric', value_name='Value')
custom_palette = ['#8c564b','#1f77b4', '#ff7f0e', '#9467bd', '#d62728', '#2ca02c']

# 使用 Seaborn 绘制折线图
sns.lineplot(x='Metric', y='Value', hue='Algorithm', data=df_melted, marker='o', palette=custom_palette)

# 设置图表标题和轴标签
plt.title('Comparison of Different Scheduling Algorithms')
plt.xlabel('del_num')
plt.ylabel('resource_utilization')

# 显示图表
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 定义列名
column_names = ['1', '2', '4', '8', '16', 'Algorithm']

# 读取数据文件
df = pd.read_csv('/ssd_data/lzw/DIF-RS/models/robust/resource_utilization.txt', sep=' ', names=column_names)

# 数据转换为适合绘图的格式
df_melted = df.melt(id_vars='Algorithm', var_name='Metric', value_name='Value')

# 自定义配色
custom_palette = ['#8c564b', '#1f77b4', '#ff7f0e', '#9467bd', '#d62728', '#2ca02c']

# 设置字体
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 16

# 设置图表大小
plt.figure(figsize=(10, 6))

# 绘制折线图
sns.lineplot(x='Metric', y='Value', hue='Algorithm', data=df_melted, marker='o', palette=custom_palette, linewidth=2)

# 设置标题和标签
plt.title('Comparison of Different Scheduling Algorithms', fontsize=18, fontweight='bold')
plt.xlabel('del_num', fontsize=16)
plt.ylabel('resource_utilization', fontsize=16)

# 设置刻度字体大小
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# 调整图例位置
plt.legend(title='Algorithm', fontsize=14, title_fontsize=16, loc='best')

# 美化网格
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图表
plt.show()
```
